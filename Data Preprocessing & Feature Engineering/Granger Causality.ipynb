```python
from statsmodels.tsa.stattools import grangercausalitytests
data = pd.read_csv(r"C:\Users\Yoon\OneDrive\한양대학교\3-2\시계열\Assignment\Project\DB\Final_1h_Feature Engineering_No Missing.csv")
# Define the target variable and independent variables
target_variable = 'cooling_power'
independent_variables = ['Wind_Temperature_Interaction','Wind_humidity_Interaction','Temp_Humidity_Interaction','Temp_Difference','Humidity_Difference','Rainfall_Humidity_Interaction','server_power','outside_temperature','outside_humidity','wind_speed','rainfall','inner_temperature','inner_humidity']

# Initialize a dictionary to store test results
results = {}

# Perform Granger Causality Tests
for var in independent_variables:
    print(f"\nGranger Causality Test: {var} -> {target_variable}")
    granger_test = grangercausalitytests(data[[target_variable, var]], maxlag=1, verbose=True)
    results[f"{var} -> {target_variable}"] = granger_test
    print("=====================================================================================")

    print(f"\nGranger Causality Test: {target_variable} -> {var}")
    granger_test = grangercausalitytests(data[[var, target_variable]], maxlag=1, verbose=True)
    results[f"{target_variable} -> {var}"] = granger_test
    print("=====================================================================================")
```
